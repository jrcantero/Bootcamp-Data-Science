## Filtrado de Datos
* Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50.
* Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.
* Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30.

```python
df[df['Total Amount'] > 50]
```

```python
df[df['Price per Unit'] < 0.5]
```

---

* Segun describe(), el 75% de los datos son mayores al valor 60, lo que trajo el 78% de los datos en la primera solicitud.
* Tambien segund describe(), los precios varian entre 25 a 500. Solicitar un precio por unidad menor a 0.5, era esperable que no arroje resultado alguno.
* Para poder obtener algun resultado, se realizo nuevamente la ultima solicitud, pero ahora solicitando valores mayores a 250

```python
# Considerando que el promedio de precio por unidad es de 180, con un minimo de 25, vamos a elegir esos datos para mostrar las filas
df[df['Price per Unit'] <= 250]
```
