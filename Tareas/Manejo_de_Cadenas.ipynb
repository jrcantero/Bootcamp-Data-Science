```python
#Ejercicio: Manipulacion de Cadenas

nombre = "Jorge Rodrigo Cantero Gutierrez"

print(nombre.upper())
print(nombre.lower())
print(len(nombre.replace(" ", "")))

nombre2 = nombre.replace("Jorge", "Alvaro")
print(nombre2)
```

```python
#Ejercicio: Lista de Nombre
input_usuario = input("Entregame una lista de nombres separados por coma (,):")
lista_nombres = input_usuario.split(",")
print(*lista_nombres, sep='\n')
```
