```python
#Definir Funcion
def suma(a: int, b: int) -> int:
  return a+b

#Llamar la funcion
resultado1 = suma(8,2)
print(f'El resultado de la funcion es: {resultado1}')

resultado2 = suma(16,26)
print(f'El resultado de la funcion es: {resultado2}')
```
