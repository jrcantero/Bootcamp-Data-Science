```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('tips')
```

```python
#SOLO PARA CONOCER ALGO MAS DE DETALLE DEL DATAFRAME Y EL TIPO DE COLUMNAS QUE TIENE CON SU TIPO DE DATOS
df.info()
```

```python
#PARA SABER EL TIPO DE DATOS QUE TIENE EL DATAFRAME
df.head(10)
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time',style='smoker', s=80)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)


#BAJE LA POSICION DE LA LEYENDA UN POCO PARA QUE SE PUDIESE VER EL PUNTO DE MAYOR TIP EN LA CENA
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles', bbox_to_anchor=(1, 0.8))

# Añadir cuadrícula
ax.grid(True, which='major', linestyle='--', linewidth=0.2, color='grey')



# Añadir anotación
ax.annotate('Mayor tip en la cena', xy=(51, 10), xytext=(30, 10),arrowprops=dict(facecolor='black', shrink=0.05))
# TOME LA DECISION DE DEJAR LA FLECHA DENTRO DE LA IMAGEN



# Añadir flecha
ax.arrow(30, 5, 10, 3, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
