```python
#Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#crear el dataframe del archivo
URL = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
df = pd.read_csv(URL)
```

```python
#REVISAMOS DATOS NULOS
df.isnull().sum()
```

```python
#ELIMINAMOS LOS NULOS DE LAS COLUMNAS QUE SOLO TIENEN UN DATO NULO, POR NO AFECTAR EL ANALISIS DE UN GRUPO DE 10 MIL REGISTROS
df.drop(df[df['Type'].isnull()].index, inplace=True)
df.drop(df[df['Content Rating'].isnull()].index, inplace=True)
```

```python
#CAMBIAMOS TIPOS DE COLUMNAS A CATEGORICAS Y LA COLUMNA DE ULTIMA ACTUALIZACION A FECHA
df['Category'] = df['Category'].astype('category')
df['Content Rating'] = df['Content Rating'].astype('category')
df['Type'] = df['Type'].astype('category')
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
#PREPARAR LAS COLUMNAS NUMERICAS Y CONVERTIRLAS
df['Installs'] = df['Installs'].astype(str).str.replace('+', '', regex=False).str.strip()
df['Installs'] = df['Installs'].str.replace(',', '', regex=False).astype(int)
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False).astype(float)
df['Size']= df['Size'].astype(str).str.replace('M', '000', regex=False).str.replace('k', '', regex=False)
df['Size'] = df['Size'].replace('Varies with device', np.nan).astype(float)
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
#MOSTRAR FILAS CON DATOS DUPLICADOS. SE USA EL SUBCONJUNTO DE APP Y CATEGORIA POR SI ACASO EXISTE ALGUN ALCANCE DE NOMBRE
df[df.duplicated(subset=['App', 'Category'], keep=False)]
```

```python
#BORRAR DATOS DUPLICADOS
df.drop_duplicates(subset=['App', 'Category'], inplace=True)
```

El siguiente paso es completar una exploración de los datos usando Python. Esto debe incluir:

```python
fig, axes = plt.subplots(figsize=(10, 6))

#ordenamos la categorias por cant. de apps
orden = df['Category'].value_counts().index


sns.countplot(x='Category', data=df, hue='Type', palette='Paired', edgecolor="black", order=orden)

axes.tick_params(axis='x', rotation=90)
axes.set_xlabel('Categorías')
axes.set_ylabel('Cantidad de aplicaciones')
axes.set_title('Cantidad de aplicaciones por categoría, categorizadas por Gratis/Pagada')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10, 6))
    
sns.countplot(x='Content Rating', data=df, hue='Type', palette='Paired', edgecolor="black")

axes.tick_params(axis='x', rotation=90)
axes.set_xlabel('Categorías')
axes.set_ylabel('Cantidad de aplicaciones')
axes.set_title('Cantidad de aplicaciones por categoría, categorizadas por Gratis/Pagada')


plt.tight_layout()
plt.show()
```

---
* LAS APLICACIONES GRATIS SON LA GRAN MAYORIA DEL DATASET
* LAS CATEGORÍAS FAMILY, GAME Y TOOLS, SON LAS QUE CONCENTRAN LA MAYOR CANTIDAD DE APLICACIONES
* CONSIDERANDO LAS APPS PAGADAS, VEMOS QUE SE CONCENTRAN EN FAMILY, Y EN SEGUNDO LUGAR ES UNA COMPETENCIA ENTRE GAME, MEDICAL, PERSONALIZATION Y TOOLS
* LA MAYOR CANTIDAD DE APPS PAGADAS SE ENCUENTRAN EN LA CATEGORIA "EVERYONE", CONSISTENTE CON LAS CATEGORIAS COMO "FAMILY" QUE CONCENTRAN LA AMYOR CANTIDAD DE APPS

```python
fig, axes = plt.subplots(figsize=(10, 6))

sns.histplot(x='Rating', data=df, kde=True, bins=35, color='Green', edgecolor='black')

#agregar una linea con el promedio del rating
promedioRating = df['Rating'].mean()
medianaRating = df['Rating'].median()

axes.axvline(promedioRating, color='red', linestyle='dashdot', label=f'Promedio: {promedioRating:.2f}')
axes.axvline(medianaRating, color='blue', linestyle='dotted', label=f'Mediana: {medianaRating:.2f}')
axes.legend(loc='upper center')

axes.tick_params(axis='x', rotation=90)
axes.set_xlabel('Rating')
axes.set_ylabel('Frecuencia')
axes.set_title('Distribución del Rating en las Apps')


plt.tight_layout()
plt.show()
```

---
* SE OBSERVA UNA AGREGACION DE PUNTAJES ENTRE 4 Y 4.5 DEL RATING DE LAS APPS
* PODRIAMOS DECIR QUE TENEMOS UN SESGO EN LOS DATOS PORQUE LA MEDIANA ES MAYOR QUE EL PROMEDIO

```python
# COMO GRAFICO MULTIVARIADO AGREGAMOS UN HEATMAP PARA VER LA CORRELACION ENTRE LAS VARIABLES NUMERICAS

plt.figure(figsize=(7,5))
plt.title('Matriz de Correlación')
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.8)

plt.tight_layout()
```

```python
plt.figure(figsize=(7,5))
plt.title('Relación entre Reviews e Installs')
sns.scatterplot(x='Reviews', y='Installs', data=df)
#se agrega linea de tendencia para mostrar la relacion vista en el heatmap
sns.regplot(x='Reviews', y='Installs', data=df, scatter=False, color='red', line_kws={'linewidth':1})

plt.ylim(0, 1100000000)
plt.xlabel('Número de Reviews')
plt.ylabel('Número de Installs')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.title('Relación entre Reviews e Installs en escala logarítmica')

sns.scatterplot(x='Reviews', y='Installs', data=df, hue='Type', palette='Paired')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('log Número de Reviews')
plt.ylabel('log Número de Installs')

plt.tight_layout()
plt.show()
```

---
* El primer grafico se hizo para mostrar la unica relacion fuertemente positiva en el heatmap junto a su linea de tendencia
* El segundo grafico en logaritmo, se realizo para observar mejor la diferencia entre las apps gratis o pagadas
* Se observa una concentracion de apps pagadas y con cantidad de reviews entre 1.000  y 100.000
