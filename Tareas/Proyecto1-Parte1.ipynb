## Carga y Preprocesamiento de Datos:
* Carga los datos del archivo CSV utilizando NumPy.
* Realiza un preprocesamiento básico para asegurarte de que los datos estén limpios y listos para su análisis.

```python
import numpy as np

def cargar_datos(ruta_archivo):
    # Carga los datos del archivo CSV utilizando NumPy
    #lei que al usar el dtype object permite cargar distintos tipos de datos
    datos = np.genfromtxt(ruta_archivo, delimiter=',', skip_header=1, dtype='object', encoding='utf-8')
    return datos

if __name__ == "__main__":
    ruta_archivo = '../data/retail_sales_dataset.csv'
    datos = cargar_datos(ruta_archivo)
    print(datos)
```

```python
#revisar si existen valores nulos en el dataset solo con numpy
Nulos = np.nan in datos
print(f"¿Hay nulos?: {Nulos}")
```

## Exploración de Datos:
* Calcula el total de ventas por categoría de producto.
* Calcula el promedio de ventas diarias por categoría de producto.
* Identifica las categorías de productos con mayores y menores ventas.

```python
#la sexta columna es la categoria de producto, el total de ventas es la ultima columna.
categorias = np.unique(datos[:, 5])  # Obtener categorías únicas
total_ventas_por_categoria = {}
for categoria in categorias:
    total_ventas = np.sum(datos[datos[:, 5] == categoria][:, -1].astype(int))  # Sumar ventas para cada categoría
    total_ventas_por_categoria[categoria] = total_ventas
```

```python
total_ventas_por_categoria
```

```python
#Calcula el promedio de ventas diarias por categoría de producto con numpy
promedio_ventas_diarias_por_categoria = {}
for categoria in categorias:
    ventas_diarias = datos[datos[:, 5] == categoria][:, -1].astype(int)  # Obtener ventas diarias para cada categoría
    promedio_ventas = np.mean(ventas_diarias)  # Calcular el promedio de ventas diarias
    promedio_ventas_diarias_por_categoria[categoria] = promedio_ventas

promedio_ventas_diarias_por_categoria
```

```python
#Identifica las categorías de productos con mayores y menores ventas.
ventas_totales = np.array(list(total_ventas_por_categoria.values()))
categorias_list = np.array(list(total_ventas_por_categoria.keys()))
categoria_mayor_ventas = categorias_list[np.argmax(ventas_totales)]
categoria_menor_ventas = categorias_list[np.argmin(ventas_totales)]
print(f"Categoría con mayores ventas: {categoria_mayor_ventas} con {total_ventas_por_categoria[categoria_mayor_ventas]} (unidad de moneda)")
print(f"Categoría con menores ventas: {categoria_menor_ventas} con {total_ventas_por_categoria[categoria_menor_ventas]} (unidad de moneda)")
```
