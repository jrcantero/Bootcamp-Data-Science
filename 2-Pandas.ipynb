# ¿Qué es Pandas?


**Pandas** es la librería de Python más usada para trabajar con datos en tabla — piénsala como un **Excel con superpoderes**.

| Excel | Pandas |
|-------|--------|
| Hoja de cálculo | DataFrame |
| Columna | Series |
| Fórmulas | Métodos Python |
| ~1M filas máximo | Millones de filas |
| Sin automatización | 100% automatizable |

> 📦 Pandas se construye sobre **NumPy** y es el corazón de todo flujo de trabajo de Data Science con Python.

# Estructuras de datos: Series


Una **Series** es una columna — una lista de valores con un índice.

```python
pd.Series(data, index=...)
```

````python
import pandas as pd

# Serie con índice automático (0, 1, 2...)
hp_pokemon = pd.Series([45, 60, 80, 45, 39], name="HP")
print(hp_pokemon)
print(f"\nTipo: {type(hp_pokemon)}")
print(f"Largo: {len(hp_pokemon)}")

# Serie con índice personalizado
hp_pokemon_named = pd.Series(
    [45, 60, 80, 45, 39],
    index=["Bulbasaur", "Ivysaur", "Venusaur", "Charmander", "Charmeleon"],
    name="HP"
)
print("\n", hp_pokemon_named)
````

# ¿Cómo funciona el filtrado?


El filtrado en pandas usa **condiciones booleanas** — igual a como aprendimos en Python, pero aplicado a columnas completas.

```python
# Paso 1: crear la condición (produce True/False por fila)
condicion = df['columna'] > valor

# Paso 2: aplicar el filtro al DataFrame
df_filtrado = df[condicion]

# O en una línea:
df_filtrado = df[df['columna'] > valor]
```

````python
import pandas as pd
df = pd.read_csv("data/pokemon.csv")

# ¿Cuántos pokémon tienen HP mayor a 100?
condicion_hp = df['HP'] > 100

print(f"Tipo de la condición: {type(condicion_hp)}")
print(f"Primeros valores:")
print(condicion_hp.head(8))
print(f"\nTotal con HP > 100: {condicion_hp.sum()}")
````

# Filtros múltiples: `&` (and), `|` (or), `~` (not)


> ⚠️ Importante: en pandas usamos `&`, `|`, `~` en lugar de `and`, `or`, `not`  
> Y cada condición debe ir entre **paréntesis**

````python
import pandas as pd
df = pd.read_csv("data/pokemon.csv")

# Pokémon de tipo Fuego Y que sean de la 1ra generación
df_fuego_gen1 = df[(df['Type 1'] == 'Fire') & (df['Generation'] == 1)]
print(f"Fuego + Gen 1: {len(df_fuego_gen1)}")
df_fuego_gen1[['Name', 'Type 1', 'Generation', 'Attack']].head()
````

````python
# Pokémon de tipo Agua O Fuego
df_agua_fuego = df[(df['Type 1'] == 'Water') | (df['Type 1'] == 'Fire')]
print(f"Agua o Fuego: {len(df_agua_fuego)}")

# Pokémon que NO son legendarios y tienen HP > 90
df_nolegen_hp = df[~df['Legendary'] & (df['HP'] > 90)]
print(f"No legendarios con HP > 90: {len(df_nolegen_hp)}")
df_nolegen_hp[['Name', 'Type 1', 'HP', 'Legendary']].head(6)
````

# ¿Por qué limpiar datos?


> *"Los científicos de datos pasan el **80% del tiempo** limpiando datos y el 20% quejándose de limpiar datos."*

Los datos reales tienen problemas:
- 🕳️ **Valores nulos** — campos vacíos o sin respuesta
- 👯 **Duplicados** — filas repetidas
- 🔤 **Tipos incorrectos** — fechas como texto, números como strings
- 🗑️ **Columnas irrelevantes** — datos que no sirven para el análisis
- ✏️ **Nombres inconsistentes** — `"Male"`, `"male"`, `"M"`, `"Hombre"`

Vamos a usar el dataset **Titanic** — tiene todos estos problemas 🚢

````python
import pandas as pd

df = pd.read_csv("data/titanic.csv")
print(f"Shape: {df.shape}")
df.head()
````

# Detectar valores nulos


````python
import pandas as pd
df = pd.read_csv("data/titanic.csv")

# ¿Cuántos nulos hay en cada columna?
nulos = df.isnull().sum()
print("Nulos por columna:")
print(nulos)

print()

# Porcentaje de nulos — más útil
pct_nulos = (df.isnull().sum() / len(df) * 100).round(1)
print("\n% de nulos por columna:")
print(pct_nulos[pct_nulos > 0])
````

````python
# Ver las filas que tienen nulos en 'Age'
df[df['Age'].isnull()][['Name', 'Sex', 'Age', 'Pclass']].head(8)
````

# Manejar valores nulos: `dropna()` y `fillna()`


| Método | Cuándo usarlo |
|--------|---------------|
| `dropna()` | Cuando hay pocos nulos y perder filas no importa |
| `fillna(valor)` | Rellenar con un valor fijo (0, "Desconocido") |
| `fillna(df['col'].mean())` | Rellenar con el promedio (variables numéricas) |
| `fillna(df['col'].mode()[0])` | Rellenar con la moda (variables categóricas) |

````python
import pandas as pd
df = pd.read_csv("data/titanic.csv")

print(f"Filas originales: {len(df)}")

# Eliminar filas donde Age es nulo
df_sin_nulos_age = df.dropna(subset=['Age'])
print(f"Después de dropna en Age: {len(df_sin_nulos_age)}")

# Mejor opción: rellenar Age con la mediana (más robusta que la media)
mediana_age = df['Age'].median()
df['Age'] = df['Age'].fillna(mediana_age)
print(f"\nMediana de edad usada: {mediana_age}")
print(f"Nulos en Age después de fillna: {df['Age'].isnull().sum()}")
````

````python
import pandas as pd
df = pd.read_csv("data/titanic.csv")

# Cabin tiene 77% de nulos — rellenar con "Desconocido"
df['Cabin'] = df['Cabin'].fillna('Desconocido')

# Embarked tiene solo 2 nulos — rellenar con la moda
moda_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(moda_embarked)

# Verificar
print("Nulos restantes:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("\n✅ Dataset limpio de nulos")
````

# Eliminar duplicados


````python
import pandas as pd
df = pd.read_csv("data/titanic.csv")

# ¿Hay filas duplicadas?
n_duplicados = df.duplicated().sum()
print(f"Filas duplicadas: {n_duplicados}")

# Eliminar duplicados (mantiene la primera ocurrencia)
df = df.drop_duplicates()
print(f"Shape después de drop_duplicates: {df.shape}")

# También podemos buscar duplicados por columnas específicas
# df.duplicated(subset=['Name', 'Age']).sum()
````

# Cambiar tipos de datos


````python
import pandas as pd
df = pd.read_csv("data/titanic.csv")

print("Tipos originales:")
print(df.dtypes)

# 'Survived' y 'Pclass' son int pero representan categorías
# .astype() para convertir
df['Survived'] = df['Survived'].astype(bool)
df['Pclass']   = df['Pclass'].astype(str)  # tratar como categoría

print("\nTipos después de conversión:")
print(df[['Survived', 'Pclass']].dtypes)

# Verificar valores únicos
print(f"\nValores en Pclass: {df['Pclass'].unique()}")
print(f"Survived es bool: {df['Survived'].dtype}")
````

# Pipeline de limpieza completo


Aplicar todos los pasos juntos — así se hace en un proyecto real:

````python
import pandas as pd

# 1. Cargar
df = pd.read_csv("data/titanic.csv")
print(f"[1] Original:       {df.shape}")

# 2. Eliminar columnas irrelevantes
df = df.drop(columns=['PassengerId', 'Ticket', 'Cabin', 'Name'])
print(f"[2] Sin columnas:   {df.shape}")

# 3. Tratar nulos
df['Age']      = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
print(f"[3] Nulos tratados: {df.isnull().sum().sum()} nulos restantes")

# 4. Eliminar duplicados
df = df.drop_duplicates()
print(f"[4] Sin duplicados: {df.shape}")

# 5. Convertir tipos
df['Survived'] = df['Survived'].astype(bool)
df['Pclass']   = df['Pclass'].astype(str)

# 6. Renombrar
df = df.rename(columns={
    'Survived': 'sobrevivio', 'Pclass': 'clase',
    'Sex': 'sexo', 'Age': 'edad', 'SibSp': 'hermanos',
    'Parch': 'hijos', 'Fare': 'tarifa', 'Embarked': 'puerto'
})

print(f"\n✅ Dataset limpio:")
print(df.dtypes)
df.head()
````
