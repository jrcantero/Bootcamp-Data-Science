**Reglas:**
- Solo letras, números y `_` — NO puede empezar con número
- Case-sensitive: `Edad` ≠ `edad` ≠ `EDAD`
- No usar palabras reservadas: `if`, `for`, `True`, `class`...
- Convención Python: `snake_case` (palabras separadas por `_`)

````python
# Variables de un perfil de usuario — igual que en una base de datos
nombre_usuario  = "maria_garcia"
edad            = 28
altura_metros   = 1.72
es_premium      = True
puntaje_riesgo  = 7.5  # score de 0 a 10

# En Data Science trabajamos con cientos de estas variables
print(f"Usuario: {nombre_usuario}")
print(f"Edad: {edad} | Altura: {altura_metros}m | Premium: {es_premium}")
print(f"Score riesgo: {puntaje_riesgo}/10")
````

# Operadores Aritméticos


| Operador | Operación | Ejemplo | Resultado | Dato curioso |
|----------|-----------|---------|-----------|--------------|
| `+` | Suma | `5 + 3` | `8` | También concatena strings |
| `-` | Resta | `9 - 4` | `5` | |
| `*` | Multiplicación | `3 * 4` | `12` | También repite strings |
| `/` | División real | `7 / 2` | `3.5` | Siempre retorna `float` |
| `//` | División entera | `7 // 2` | `3` | Descarta el decimal |
| `%` | Módulo (resto) | `7 % 2` | `1` | Clave para pares/impares |
| `**` | Potencia | `2 ** 10` | `1024` | En ML: regularización L2 |

````python
# Caso real: calcular métricas de negocio
ventas_mes     = 8_500_000
ventas_anterior = 7_200_000
costo_total    = 3_100_000

crecimiento_pct = (ventas_mes - ventas_anterior) / ventas_anterior * 100
margen          = (ventas_mes - costo_total) / ventas_mes * 100
ticket_promedio = ventas_mes / 1_243  # número de transacciones

print(f"Crecimiento MoM:  {crecimiento_pct:.1f}%")
print(f"Margen bruto:     {margen:.1f}%")
print(f"Ticket promedio:  ${ticket_promedio:,.0f}")

# Módulo — muy usado en programación
for n in range(1, 11):
    tipo = "par" if n % 2 == 0 else "impar"
    print(f"{n}: {tipo}")
````

# Listas — Indexing y Slicing


Las listas son la **estructura más usada en Python**. En pandas, cada columna es internamente una lista.

```
 lista = ["a", "b", "c", "d", "e"]
índice =   0    1    2    3    4     (positivos →)
índice =  -5   -4   -3   -2   -1    (← negativos)
```

````python
# Simula la columna "temperatura" de un dataset meteorológico
temperaturas = [22.1, 18.5, 25.3, 30.1, 15.8, 28.7, 19.2]

print(f"Primera medición:  {temperaturas[0]}")
print(f"Última medición:   {temperaturas[-1]}")
print(f"Rango días 2-4:    {temperaturas[1:4]}")
print(f"Cada 2 días:       {temperaturas[::2]}")
print(f"Invertida:         {temperaturas[::-1]}")
print(f"Total mediciones:  {len(temperaturas)}")

# Operaciones estadísticas básicas (sin librerías)
print(f"\nPromedio: {sum(temperaturas)/len(temperaturas):.1f}°C")
````

# Listas — Métodos útiles


````python
alumnos = ["Ana", "Luis", "María"]

alumnos.append("Carlos")       # agrega al final
alumnos.insert(1, "Sofía")     # inserta en posición 1
alumnos.remove("Luis")         # elimina por valor
alumnos.sort()                 # ordena in-place (modifica la lista)

print(alumnos)
print(f"Total alumnos: {len(alumnos)}")

# Verificar pertenencia — MUY eficiente
print("Ana" in alumnos)        # True
print("Pedro" in alumnos)      # False

# List comprehension — la forma pythónica
notas = [6.8, 4.2, 5.5, 7.0, 3.8, 6.1]
aprobados = [n for n in notas if n >= 4.0]
print(f"\nAprobados: {aprobados}")
print(f"Reprobados: {len(notas) - len(aprobados)}")
````

# Diccionarios

# Strings — Métodos que usarás en limpieza de datos


En Data Science, el texto sucio es la norma. Estos métodos serán tus aliados en pandas:

````python
# Dato tal como llega de un formulario web mal limpiado
entrada = "  Juan PÉREZ  ,  Santiago  , CHILE  "

# Limpieza paso a paso (pipeline de texto)
limpio = entrada.strip()           # quita espacios extremos
limpio = limpio.lower()            # normaliza mayúsculas
partes = limpio.split(",")         # divide en lista
partes = [p.strip() for p in partes]  # limpia cada parte

nombre   = partes[0].title()  # .title() capitaliza cada palabra
ciudad   = partes[1].title()
pais     = partes[2].upper()

print(f"Nombre: {nombre}")
print(f"Ciudad: {ciudad}")
print(f"País:   {pais}")

# Otras operaciones útiles
email = "usuario@correo.com"
print(f"\n¿Es email válido? {'@' in email and '.' in email.split('@')[1]}")
print(f"Dominio: {email.split('@')[1]}")
print(f"Largo: {len(email)} caracteres")
````

# f-strings — El estándar para formatear texto


Las f-strings (Python 3.6+) son la forma más legible y eficiente de insertar variables en texto:
```python
f"texto {variable} texto {expresión:.2f}"
```

````python
nombre  = "María"
ventas  = 1_234_567.89
tasa    = 0.1478
activo  = True

# Formato de números
print(f"Ventas:    ${ventas:,.2f}")         # separador de miles + 2 decimales
print(f"Tasa:      {tasa:.1%}")             # porcentaje con 1 decimal
print(f"Científico:{ventas:.2e}")           # notación científica
print(f"Entero:    {ventas:.0f}")           # sin decimales

# Alineación de texto (útil para tablas en consola)
for producto, precio in [("Laptop", 899990), ("Mouse", 15990), ("Teclado", 49990)]:
    print(f"{producto:<12} ${precio:>10,.0f}")  # < izq, > der

# Expresiones dentro de f-string
print(f"\n{nombre} {'✅ activo' if activo else '❌ inactivo'}")
print(f"Con IVA: ${ventas * 1.19:,.2f}")
````

# Operadores de Comparación y Lógicos


Todos retornan `True` o `False` — son la base de **cualquier filtro en Data Science**.

| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|-----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Distinto de | `5 != 3` | `True` |
| `>` / `<` | Mayor / Menor | `7 > 3` | `True` |
| `>=` / `<=` | Mayor-igual / Menor-igual | `5 >= 5` | `True` |
| `and` | Ambos verdaderos | `T and F` | `False` |
| `or` | Al menos uno verdadero | `T or F` | `True` |
| `not` | Negación | `not True` | `False` |
| `in` | Pertenencia | `3 in [1,2,3]` | `True` |

> 🎯 En pandas: `df[df['edad'] > 30 & df['activo'] == True]` — esto es exactamente lo que aprenderás.

````python
# Operadores en acción — simula un sistema de elegibilidad de crédito
ingreso       = 850_000
deuda_actual  = 200_000
historial     = "bueno"
meses_trabajo = 18

# Condiciones individuales
ingreso_suficiente  = ingreso >= 600_000
deuda_manejable     = deuda_actual / ingreso < 0.40
buen_historial      = historial in ["bueno", "excelente"]
estabilidad_laboral = meses_trabajo >= 12

print(f"Ingreso suficiente:    {ingreso_suficiente}")
print(f"Deuda manejable:       {deuda_manejable}  ({deuda_actual/ingreso:.0%} del ingreso)")
print(f"Buen historial:        {buen_historial}")
print(f"Estabilidad laboral:   {estabilidad_laboral}")

elegible = ingreso_suficiente and deuda_manejable and buen_historial and estabilidad_laboral
print(f"\n→ Elegible para crédito: {elegible}")
````

# Condicional `if / elif / else`


```python
if <condición_1>:
    # bloque si condición_1 es True
elif <condición_2>:
    # bloque si condición_2 es True (y condición_1 fue False)
else:
    # bloque si ninguna condición fue True
```

> ⚠️ La **indentación** (4 espacios) NO es opcional en Python — define los bloques

````python
# Caso real: clasificación de riesgo crediticio (usado en fintech)
score = 720   # prueba con distintos valores: 300, 580, 670, 720, 800

if score >= 750:
    nivel = "Excelente"
    tasa  = 0.08
    monto_max = 50_000_000
elif score >= 700:
    nivel = "Muy bueno"
    tasa  = 0.12
    monto_max = 30_000_000
elif score >= 650:
    nivel = "Bueno"
    tasa  = 0.18
    monto_max = 15_000_000
elif score >= 600:
    nivel = "Regular"
    tasa  = 0.28
    monto_max = 5_000_000
else:
    nivel = "Bajo"
    tasa  = None
    monto_max = 0

print(f"Score:        {score}")
print(f"Nivel:        {nivel}")
if monto_max > 0:
    print(f"Tasa:         {tasa:.0%} anual")
    print(f"Monto máximo: ${monto_max:,.0f}")
else:
    print("❌ Crédito no aprobado")
````

# Estructura `for`


Itera un número **fijo** de veces sobre cualquier secuencia.

```python
for variable in secuencia:
    # instrucciones
```

Funciona con: listas, strings, rangos, diccionarios, archivos, DataFrames...

````python
# Caso real: evaluar múltiples modelos (preview de lo que harás en ML)
modelos = [
    {"nombre": "Regresión Lineal",  "accuracy": 0.742, "tiempo_s": 0.3},
    {"nombre": "KNN",               "accuracy": 0.815, "tiempo_s": 1.2},
    {"nombre": "Random Forest",     "accuracy": 0.891, "tiempo_s": 4.7},
    {"nombre": "XGBoost",           "accuracy": 0.903, "tiempo_s": 6.1},
]

mejor = max(modelos, key=lambda m: m["accuracy"])

print(f"{'MODELO':<22} {'ACCURACY':>10} {'TIEMPO':>10}")
print("-" * 46)

for modelo in modelos:
    marca = " ⭐" if modelo == mejor else ""
    print(f"{modelo['nombre']:<22} {modelo['accuracy']:>9.1%} {modelo['tiempo_s']:>8.1f}s{marca}")

print(f"\nMejor modelo: {mejor['nombre']} con {mejor['accuracy']:.1%}")
````

# Estructura `while`


Ejecuta un bloque **mientras** una condición sea `True`. Útil cuando no sabemos de antemano cuántas veces iterar.

> ⚠️ Siempre asegúrate de que la condición eventualmente sea `False` — de lo contrario es un **loop infinito**.

- `break` — sale del loop inmediatamente
- `continue` — salta a la siguiente iteración

````python
# Caso real: simular descenso de gradiente (algoritmo de optimización en ML)
# En cada iteración la "pérdida" debe disminuir hasta converger

import random
random.seed(42)

perdida     = 1.0     # pérdida inicial (error del modelo)
iteracion   = 0
max_iter    = 1000
tolerancia  = 0.01   # converge cuando la pérdida cae bajo este umbral

print(f"{'Iteración':>10} | {'Pérdida':>10}")
print("-" * 25)

while perdida > tolerancia and iteracion < max_iter:
    # Simula una reducción aleatoria de la pérdida
    reduccion = random.uniform(0.05, 0.15)
    perdida  *= (1 - reduccion)
    iteracion += 1

    if iteracion % 5 == 0:  # imprime cada 5 iteraciones
        print(f"{iteracion:>10} | {perdida:>10.4f}")

print("-" * 25)
print(f"\n✅ Convergió en {iteracion} iteraciones — Pérdida final: {perdida:.4f}")
````

````python
# break y continue — control fino del loop
print("Buscando el primer número divisible por 7 entre 50 y 100:")
for n in range(50, 101):
    if n % 7 != 0:
        continue        # salta los no divisibles
    print(f"Encontrado: {n}")
    break               # para al primer resultado
````

````python
# ▶️ CELDA INTERACTIVA — ejecutar manualmente (no incluir en "Run All")
# Validación de input con while — patrón muy común en formularios y apps
intentos_max = 3
intentos     = 0

while intentos < intentos_max:
    nota_str = input(f"  Ingresa una nota (1.0-7.0) — intento {intentos+1}/{intentos_max}: ")
    try:
        nota = float(nota_str)
        if 1.0 <= nota <= 7.0:
            print(f"  ✅ Nota registrada: {nota}")
            break
        print("  ⚠️  La nota debe estar entre 1.0 y 7.0")
    except ValueError:
        print("  ⚠️  Ingresa un número válido")
    intentos += 1
else:
    print("  ❌ Demasiados intentos fallidos")
````

# List Comprehensions *(Pythónico)*


Forma compacta y eficiente de crear listas. Verás esto constantemente en código de Data Science.

```python
nueva_lista = [expresión for elemento in iterable if condición]
```

````python
# Comparación: for tradicional vs list comprehension
temperaturas = [22.1, -3.5, 18.0, 35.2, 0.1, 28.7, -1.2, 15.3]

# Forma tradicional
positivas_trad = []
for t in temperaturas:
    if t > 0:
        positivas_trad.append(t)

# List comprehension (equivalente, más conciso)
positivas = [t for t in temperaturas if t > 0]
celsius_a_fahrenheit = [(t * 9/5) + 32 for t in positivas]

print(f"Temperaturas positivas: {positivas}")
print(f"En Fahrenheit:          {[f'{f:.1f}' for f in celsius_a_fahrenheit]}")

# Caso DS: normalizar una columna (Min-Max Scaling)
datos = [10, 25, 8, 42, 15, 37, 5]
min_val, max_val = min(datos), max(datos)
normalizados = [(x - min_val) / (max_val - min_val) for x in datos]
print(f"\nDatos originales:  {datos}")
print(f"Normalizados [0,1]: {[round(x,2) for x in normalizados]}")
````

## 🏁 Actividad Final — Semana 1: Sistema de Notas

### Problema

Construye un **sistema de gestión de notas** que haga lo siguiente:

1. **Ingresa** 5 notas (de 1.0 a 7.0) validando que estén en rango
2. **Calcula** sin usar `max()`/`min()` directamente:
   - Promedio
   - Nota más alta y más baja (usando un `for`)
   - Cuántas notas son aprobadas (≥ 4.0)
3. **Clasifica** el rendimiento:
   - Promedio ≥ 5.5 → `"Destacado 🌟"`
   - Promedio ≥ 4.0 → `"Aprobado ✅"`
   - Promedio < 4.0 → `"Reprobado ❌"`
4. **Imprime** un reporte formateado con f-strings

**Bonus:** muestra una barra de progreso ASCII del promedio (ej: `████░░░ 5.2/7.0`)

````python
# Actividad Final — Sistema de Notas
print("=" * 42)
print("  SKILLNEST — Sistema de Registro de Notas")
print("=" * 42)

notas = []
for i in range(1, 6):
    while True:
        try:
            nota = float(input(f"  Nota {i}/5 (1.0 – 7.0): "))
            if 1.0 <= nota <= 7.0:
                notas.append(nota)
                break
            print("  ⚠️  Debe estar entre 1.0 y 7.0")
        except ValueError:
            print("  ⚠️  Ingresa un número válido")

# Calcular estadísticas con for (sin max/min built-in)
mayor     = notas[0]
menor     = notas[0]
aprobadas = 0

for n in notas:
    if n > mayor:
        mayor = n
    if n < menor:
        menor = n
    if n >= 4.0:
        aprobadas += 1

promedio = sum(notas) / len(notas)

# Clasificación
if promedio >= 5.5:
    veredicto = "Destacado 🌟"
elif promedio >= 4.0:
    veredicto = "Aprobado ✅"
else:
    veredicto = "Reprobado ❌"

# Barra de progreso ASCII (bonus)
barras     = int((promedio / 7.0) * 20)
barra_str  = "█" * barras + "░" * (20 - barras)

# Reporte final
print()
print("=" * 42)
print(f"  Notas:       {notas}")
print(f"  Promedio:    {promedio:.2f}  [{barra_str}]")
print(f"  Más alta:    {mayor}")
print(f"  Más baja:    {menor}")
print(f"  Aprobadas:   {aprobadas}/5")
print(f"  Veredicto:   {veredicto}")
print("=" * 42)
````
